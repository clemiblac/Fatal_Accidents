```python
import pandas as pd
import chardet
```

```python
accidents=pd.read_csv("Resources/ACCIDENT.csv")
accidents.head(5)
```

```python
accidents.columns
```

```python
accidents1=accidents[['STATE', 'ST_CASE','PERMVIT','COUNTY', 'CITY', 'DAY', 'MONTH',
       'YEAR', 'DAY_WEEK', 'HOUR', 'MINUTE', 'RUR_URB','ROUTE','LATITUDE','LONGITUD',
       'HARM_EV', 'MAN_COLL','LGT_COND', 'REL_ROAD','WEATHER', 'SCH_BUS',
        'CF1', 'CF2', 'CF3','FATALS','DRUNK_DR']]
```

```python
vehicle=pd.read_csv("Resources/VEHICLE.csv",engine='python')
vehicle.head(5)
```

```python
vehicle.columns
```

```python
vehicle1=vehicle[['ST_CASE','MAKE','MODEL','BODY_TYP','MOD_YEAR','M_HARM','DEATHS']]
vehicle1
```

```python
person=pd.read_csv("Resources/PERSON.csv")
person.head(5)
```

```python
person.columns
```

```python
person1=person[['ST_CASE','PER_NO','AGE', 'SEX','DEATH_DA', 'DEATH_MO', 'DEATH_YR']]
```

```python
df1=pd.merge(accidents1,vehicle1,on='ST_CASE')
```

```python
data=pd.merge(df1,person1,on='ST_CASE')
```

```python
data.columns
```

```python
data.head(5)
```

```python
import json
```
